# Problem 1 (35 points)

Solve the following LP model using the graphical method.


$Min(Z) = 30X + 32Y$

subject to:
* $9X + 9Y \leq 720$
* $X + 2Y \geq 40$
* $X \geq 10$
* $Y \geq 40$
* $X,Y \geq 0$



```python
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Set up the figure
plt.figure(figsize=(6, 6))
plt.axis([0, 100, 0, 100])
plt.xlabel('X')
plt.ylabel('Y')

# Constraint: 9X + 9Y <= 720
x = np.array([0, 80])
y = np.array([80, 0])
plt.plot(x, y, 'gold', lw=2, label='9X + 9Y <= 720')

plt.fill_between(
    [0, 80, 100],
    [80, 0, 0],
    [100, 100, 100],
    color='gold',
    alpha=0.15
)

# Constraint: X + 2Y >= 40
x = np.array([0, 40])
y = np.array([20, 0])
plt.plot(x, y, color='magenta', lw=2, label='X + 2Y >= 40')

plt.fill_between(
    [0, 40],
    [0, 0],
    [20, 0],
    color='magenta',
    alpha=0.15
)

# Constraint: X >= 10
x = np.array([10, 10])
y = np.array([0, 100])
plt.plot(x, y, 'maroon', lw=2, label='X >= 10')

plt.fill_between(
    [0, 10],
    [0, 0],
    [100, 100],
    color='maroon',
    alpha=0.15
)

# Constraint: Y >= 40
x = np.array([0, 100])
y = np.array([40, 40])
plt.plot(x, y, 'indigo', lw=2, label='Y >= 40')

plt.fill_between(
    [0, 100],
    [0, 0],
    [40, 40],
    color='indigo',
    alpha=0.15
)

# For X, Y >= 0 (axes), as they are covered by other constraints

# Draw the minimal line 30X + 32Y = 1580
x = np.array([0, 1580 / 30])
y = np.array([1580 / 32, 0])
plt.plot(x, y, 'teal', lw=3, label='30X + 32Y = 1580')

# Objective function: minimize 30X + 32Y
objective_coefficients = np.array([30, 32])

print('**** Point 1 ****')
A = np.array([[0, 1], [9, 9]])
b = np.array([40, 720])
x = np.linalg.solve(A, b)
print('x:', x)
value1 = np.dot(x, objective_coefficients)
print('value1:', value1)

print('**** Point 2 ****')
A = np.array([[1, 0], [9, 9]])
b = np.array([10, 720])
x = np.linalg.solve(A, b)
print('x:', x)
value2 = np.dot(x, objective_coefficients)
print('value2:', value2)

print('**** Point 3 ****')
A = np.array([[1, 0], [0, 1]])
b = np.array([10, 40])
x = np.linalg.solve(A, b)
print('x:', x)
value3 = np.dot(x, objective_coefficients)
print('value3:', value3)

print('The minimum value of the function 30X + 32Y is', min(value1, value2, value3))

#Show the legend
plt.legend()

#Show the plot
plt.show()
```

# Problem 2 (35 points)

Solve the following LP model using the graphical method.

$Max(Z) = 15X + 10Y$

subject to:
* $3X + 5Y \leq 45$
* $4X + Y \leq 20$
* $X,Y \geq 0$ (nonnegativity)



```python

# Set up the figure
plt.figure(figsize=(6, 6))
plt.axis([0, 30, 0, 30])
plt.xlabel('X')
plt.ylabel('Y')

# Constraint: 3X + 5Y <= 45
x = np.array([0, 15])
y = np.array([9, 0])
plt.plot(x, y, 'gold', lw=2, label='3X + 5Y <= 45')

plt.fill_between(
    [0, 15, 30],
    [9, 0, 0],
    [30, 30, 30],
    color='gold',
    alpha=0.15
)

# Constraint: 4X + Y <= 20
x = np.array([0, 5])
y = np.array([20, 0])
plt.plot(x, y, color='magenta', lw=2, label='4X + Y <= 20')

plt.fill_between(
    [0, 5, 30],
    [20, 0, 0],
    [30, 30, 30],
    color='magenta',
    alpha=0.15
)

# For X, Y >= 0 (nonnegativity), as they are covered by other constraints

# Draw the objective function line 15X + 10Y = Z
#x = np.array([0, 30])
#y = (15 * x) / 10
x = np.array([0, 119.12 / 15])
y = np.array([119.12 / 10, 0])
plt.plot(x, y, 'teal', lw=3, label='15X + 10Y = 119.12')

# Objective function: Minimize 15X + 10Y
objective_coefficients = np.array([15, 10])

print('**** Point 1 ****')
A = np.array([[0, 1], [1, 0]])
b = np.array([0, 0])
x = np.linalg.solve(A, b)
print('x:', x)
value1 = np.dot(x, objective_coefficients)
print('value1:', value1)

print('**** Point 2 ****')
A = np.array([[1, 0], [3, 5]])
b = np.array([0, 45])
x = np.linalg.solve(A, b)
print('x:', x)
value2 = np.dot(x, objective_coefficients)
print('value2:', value2)

print('**** Point 3 ****')
A = np.array([[4, 1], [0, 1]])
b = np.array([20, 0])
x = np.linalg.solve(A, b)
print('x:', x)
value3 = np.dot(x, objective_coefficients)
print('value3:', value3)

print('**** Point 4 ****')
A = np.array([[4, 1], [3, 5]])
b = np.array([20, 45])
x = np.linalg.solve(A, b)
print('x:', x)
value4 = np.dot(x, objective_coefficients)
print('value4:', value4)

print('The maximum value of the function 15X + 10Y is', round(max(value1, value2, value3, value4), 2))

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
